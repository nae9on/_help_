```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
# simple line plot
slope = [1,2,3]
c = 0
x = np.arange(100)

for m in slope:
    y = m*x+c
    # squeeze returns a rank 1 array which is good for plotting
    plt.plot(np.squeeze(y), label = "slope="+str(m))

plt.xlabel('X')
plt.ylabel('Y')
legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
#plt.show()
```

```python
# simple scatter plto
theta = np.arange(0,360,1)
R = 1; x = R*np.cos(theta*math.pi/180); y = R*np.sin(theta*math.pi/180)
plt.scatter(x, y, c=y, s=40, cmap=plt.cm.Spectral);
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.gca().set_aspect('equal', adjustable='box')
#plt.draw()
```
