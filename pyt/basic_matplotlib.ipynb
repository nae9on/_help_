```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# simple line plot
slope = [1,2,3]
c = 0
x = np.arange(100)
for m in slope:
    y = m*x+c
    plt.plot(np.squeeze(y), label = "slope="+str(m))
    plt.ylabel('Y')
plt.xlabel('X')

legend = plt.legend(loc='upper center', shadow=True)

frame = legend.get_frame()

frame.set_facecolor('0.90')

plt.show()
```
